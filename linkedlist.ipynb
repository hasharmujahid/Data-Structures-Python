````python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


class LinkedList:
    def __init__(self):
        self.head = None




def __str__(self):
    retstr= '['
    tempnode = self.head
    while tempnode is not None:
        retstr += str(tempnode.data) + ', '
        tempnode = tempnode.next

    retstr = retstr.rstrip(', ')
    retstr += ']'
    return retstr

# Str Dunder Function       
LinkedList.__str__ = __str__

````

# PUSH FUNCATION


````python
# Push Operation Logic
def push(self,data):
    new_node = Node(data)

    if self.head is None:
        self.head = new_node
        return
    last = self.head
    while last.next is not None:
        last = last.next
    
    last.next = new_node
````

````python

LinkedList.push = push
````

````python
list = LinkedList()

list.push(1)
list.push(2)
list.push(3)
list.push(4)
list.push(5)
list.push(6)
list.push(7)
list.push(9)
list.push(10)

print(list)
````

# POP OPERATION


We Have 2 Cases 






````python

def pop(self):
    last = self.head
    
    if last.next is None:
        print('Cannot Pop the Value, No node present')
        return

    while last.next is not None:
        prev = last
        last = last.next
    val = last.data
    prev.next = None
    return val


LinkedList.pop = pop


list.pop()
print(list)
````

# INSERT OPERATION

We also have 2 cases in Insertion

> Insert At the head or 0




> Explanation Case 1

Sure, I can explain this with a diagram. This code is for inserting a new node at the beginning of a linked list.

Here's the situation before the insertion:

```
head --> Node1 --> Node2 --> Node3
```

`self.head` is a reference to the first node in the list (Node1).

When you want to insert a new node (`new_node`) at the beginning of the list (index 0), you need to do two things:

1. Set the `next` attribute of `new_node` to refer to the current head of the list. This is what `new_node.next = self.head` does. After this step, the situation looks like this:

```
head --> Node1 --> Node2 --> Node3
  |
new_node
  |
  V
Node1 --> Node2 --> Node3
```

2. Update `self.head` to refer to `new_node`. This is what `self.head = new_node` does. After this step, the situation looks like this:

```
head
  |
new_node --> Node1 --> Node2 --> Node3
```

So now, `new_node` is the first node in the list, and the original first node (Node1) is the second node in the list.


> Insert at any operation except 0





````python
def insert(self,data, index):
    new_node = Node(data=data)

    # case 1 check if index is zero if it is then execute case 1 logic

    if index == 0:
        new_node.next = self.head
        self.head = new_node
        return
    
    counter = 0
    temp = self.head

    while temp is not None and counter < index:
        prev = temp
        temp = temp.next
        counter +=1
    prev.next = new_node
    new_node.next = temp

LinkedList.insert = insert

````

````python
list.insert(8,7)
print(list)
list.insert(0,0)
print(list)
````

# REMOVE OPERATIONS

Same cases as Insertion Operation

> Remove at index 0 




> Remove at anyother index





````python
def remove(self, val):

    # Case 1 
    temp = self.head

    if temp is not None:
        if temp.data == val:
            self.head = temp.next # changed the current head to the next node 
            temp = None

            return

    while temp is not None and temp.data != val:
        prev = temp
        temp = temp.next 
    
    prev.next = temp.next
    temp = None

    if temp is None:
        return
    
LinkedList.remove = remove

list.remove(5)
print(list)

````
